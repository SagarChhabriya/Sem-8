### 1. Simple Imputer
### 2. KNN Imputer

```python
from sklearn.impute import SimpleImputer
from sklearn.impute import KNNImputer
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
    "a": [np.nan,2, 3, 4,5],
    "b": [1,np.nan, 3,4,5],
    "c": [1,2,np.nan, 4,5],
    "d": [1,2,3,np.nan,5],
    "e": [1,2,3,4,np.nan]
})
```

```python
df
```

```python
simple_imputer = SimpleImputer(strategy="mean")

simple_imputed_df = simple_imputer.fit_transform(df)
```

```python
simple_imputed_df
```

```python
knn_imputer = KNNImputer(n_neighbors=3)
```

```python
knn_imputer.fit_transform(df)
```
